# 1. Data visualization on yield

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv('./temp_yield.csv')
```

```python
df.sample(n=5)
```

```python
df.head()
```

```python
fr = df[df['Country'] == 'France']
fr[['Country', 'year', 'avg_temp', 'HWN', 'HWM', 'CWN', 'CWM']]
```

```python
df.describe(include='all')
```

```python
avgyield = df.groupby('Country')['Yield_Value'].mean().reset_index()

avgtemp = df.groupby('Country')['avg_temp'].mean().reset_index()

avgdata = pd.merge(avgyield, avgtemp, on='Country')
avgdata.tail(20)
```

```python
fig_yield = px.choropleth(
    avgdata,
    locations="Country",
    locationmode="country names",
    color="Yield_Value",
    hover_name="Country",
    title="Average wheat yield by country",
    labels={"Yield_Value": "Yield value (g/ha)"}
)

fig_yield.show()
```

```python
yieldstats = df.groupby('Country')['Yield_Value'].describe()

yieldstats.sort_values('mean', ascending=False)
```

```python
major = df[df['Country'].isin(['India', 'France', 'Canada', 'USSR', 'Russian Federation', 'United States of America', 'Australia', 'Ukraine', 'China'])]

fig_box = px.box(
    major,
    x='Country',
    y='Yield_Value',
    title='Wheat yield distribution for major producers',
    points="all"
)

fig_box.show()
```

```python
avgyield = df.groupby('Country')['Yield_Value'].mean().reset_index()
avgyield['Major_Producer'] = avgyield['Country'].isin(['France', 'Canada', 'USSR', 'Russian Federation', 
                                                       'United States of America', 'Australia', 'Ukraine', 'China'])

fig_scatter = px.scatter(
    avgyield,
    x='Country',
    y='Yield_Value',
    color='Major_Producer',
    title='Comparison of average wheat yields: major producers vs. other countries'
)

fig_scatter.show()
```

```python
fig_line = px.line(
    major,
    x='year',
    y='Yield_Value',
    color='Country',
    title='Wheat yield over time for major producers',
    labels={"Yield_Value": "Yield value (g/ha)", 'year':'Year'},
    category_orders={'Country': ['China', 'India', 'United States of America', 'Russian Federation',
                                  'USSR', 'France', 'Australia', 'Canada', 'Ukraine']}

)

average_yield = major.groupby('year')['Yield_Value'].mean().reset_index()

fig_line.add_trace(
    go.Scatter(
        x=average_yield['year'],
        y=average_yield['Yield_Value'],
        mode='lines',
        name='Average yield',
        line=dict(dash='dot', color='black') 
    )
)

fig_line.show()
```

```python
fig_temp = px.choropleth(
    avgdata,
    locations="Country",
    locationmode="country names",
    color="avg_temp",
    hover_name="Country",
    title="Average temperature by country",
    labels={"avg_temp": "Average temperature (°C)"}

)

fig_temp.show()
```

```python
avgyield_year = df.groupby('year')['Yield_Value'].mean().reset_index()

fig_AY = px.line(
    avgyield_year,
    x='year',
    y='Yield_Value',
    title='Worldwide average wheat yield (1961 - 2022)',
    labels={'Yield_Value': 'Average yield (g/ha)', 'year': 'Year'}
)

fig_AY.show()
```

```python
avgextreme = df.groupby('year')[['Tmax', 'Tmin']].mean().reset_index()

fig_ex = px.line(
    avgextreme,
    x='year',
    y=['Tmax', 'Tmin'],
    title='Evolution of temperature extremes over the years',
    labels={'value': 'Temperature (°C)', 'year': 'Year'}
)

fig_ex.show()
```

```python
fig_yield_temp = px.scatter(
    df,
    x='avg_temp',
    y='Yield_Value',
    color='Country',
    title='Wheat yield vs Average temperature',
    labels={'avg_temp': 'Average temperature (°C)', 'Yield_Value': 'Yield value'}
)

fig_yield_temp.show()
```

```python
fig_yield_hist = px.histogram(
    df,
    x='Yield_Value',
    nbins=50,
    title='Distribution of wheat yield values',
    labels={'Yield_Value': 'Yield value'}
)

fig_yield_hist.show()
```

# 2. Data visualization on production

```python
df = pd.read_csv('./FAOSTAT_data.csv')
```

## A. Cleaning the dataset

```python
def check(df):
    c = []
    for country in df.Area.value_counts().items():
        for year, y in df[df.Area == country[0]].Year.value_counts().items():
            if y == 1:
                c.append(country[0])
            else:
                pass
    return c

rawcountries = check(df)
cleaned = dict()

for e in rawcountries:
    if e in cleaned:
        cleaned[e] += 1
    else:
        cleaned[e] = 1

for key in cleaned.keys():
    if key == 'China':
        pass
    else:
        df = df[df.Area != key]

for col in ("Domain Code", "Domain", "Area Code (M49)", "Element Code", "Item Code (CPC)", "Item", "Year Code", "Flag",
         "Flag Description", "Note"):
    df.drop(col, axis=1, inplace=True)

df = df[df.Element == "Production"]
df = df.reset_index(drop=True)
df.drop("Element", axis=1, inplace=True)
```

## B. Data visualization

```python
avgprod = df.groupby('Area')['Value'].mean().reset_index()
avgprod.columns = ['Country', 'Average_Production']

fig_map = px.choropleth(
    avgprod,
    locations='Country',
    locationmode='country names',
    color='Average_Production',
    hover_name='Country',
    title='Average wheat production by country',
    labels={'Average_Production': 'Average production (t)'}
)

fig_map.show()
```

```python
avgprod = df.groupby('Year')['Value'].mean().reset_index()
avgprod.columns = ['Year', 'Average_Production']

fig_avg_year = px.line(
    avgprod,
    x='Year',
    y='Average_Production',
    title='Average wheat production (1961 - 2022)',
    labels={'Average_Production': 'Average production (tonnes)'}
)

fig_avg_year.show()
```

```python
prodyear = df.groupby('Year')['Value'].sum().reset_index()
prodyear.columns = ['Year', 'Total_Production']

fig_line = px.line(
    prodyear,
    x='Year',
    y='Total_Production',
    title='Evolution of global wheat production (1961 - 2022)',
    labels={'Total_Production': 'Total production (tonnes)'}
)

fig_line.show()
```

```python
major = df[df['Area'].isin(['France', 'Canada', 'USSR', 'Russian Federation', 'United States of America', 'Australia', 'Ukraine', 'China'])]

fig_box = px.box(
    major,
    x='Area',
    y='Value',
    title='Wheat production distribution for major producers',
    points="all"
)

fig_box.show()
```

```python
major = df[df['Area'].isin(['France', 'Canada', 'USSR', 'Russian Federation', 'United States of America', 'Australia', 'Ukraine', 'China'])]
```

```python
fig_area = px.area(
    major,
    x='Year',
    y='Value',
    color='Area',
    title='Wheat production over time for major producers',
    labels={'Value':'Value (tonnes)'},
    category_orders={'Area': ['China', 'India', 'United States of America', 'Russian Federation',
                                  'USSR', 'France', 'Australia', 'Canada', 'Ukraine']}
)

fig_area.show()
```

```python
major_prod = ['China', 'India', 'United States of America', 'Russian Federation',
                   'USSR', 'France', 'Australia', 'Canada', 'Ukraine']

minor = df[~df['Area'].isin(major_prod)]
minorsum = minor.groupby('Year')['Value'].sum().reset_index()
minorsum['Area'] = 'Other Countries'

major_filtered = major[major['Area'].isin(major_prod)]

combined_df = pd.concat([major_filtered, minorsum])

print(combined_df.tail())

fig_area = px.area(
    combined_df,
    x='Year',
    y='Value',
    color='Area',
    title='Wheat Production Over Time for Major Producers and Other Countries',
    labels={'Value': 'Value (tonnes)'},
    category_orders={'Area': ['China', 'India', 'United States of America', 'Russian Federation',
                              'USSR', 'France', 'Australia', 'Canada', 'Ukraine', 'Other Countries']}
)

fig_area.show()
```

```python
fig_yield_hist = px.histogram(
    df,
    x='Value',
    nbins=50,
    title='Distribution of wheat production values',
    labels={'Value': 'Production value (tonnes)'}
)

fig_yield_hist.show()
```
